program to check whether a string is palindrome or not

```python
def isPalindrome(str):
  for i in range(0,int(len(str)/2)):
    if str[i] != str[len(str)-i-1]:
      return False
    return True
s = input("Enter a string: ")
ans = isPalindrome(s)
if (ans):
  print("The string is Palindrome")
else:
  print("The string is not Palindrome")
```

program to print the largest and the second largest element in an array

```python

def lar_seclar(arr):
  if len(arr) < 2:
    return None, None
  largest = max(arr)
  second_largest=float('-inf')
  for i in arr:
    if i != largest and i > second_largest:
      second_largest=i
  if second_largest == float('-inf'):
    second_largest = None
  return largest, second_largest

arr =  [7,9,0,5,7,2]
largest, second_largest = lar_seclar(arr)
print(f"The largest is: {largest}")
print(f"The second largest is: {second_largest}")
```

Write a program that reads two 2D metrices from the console, verifies if metrics
multiplication is possible or not. Then multiplies the metrices and prints the 3rd metrics.

```python
def read_matrix(rows, cols):
    """Reads a matrix of given dimensions from the console."""
    matrix = []
    print(f"Enter the elements of a {rows}x{cols} matrix:")
    for i in range(rows):
        row = list(map(int, input().split()))
        if len(row) != cols:
            print(f"Error: You must enter {cols} numbers per row.")
            return None
        matrix.append(row)
    return matrix

def can_multiply_matrices(A, B):
    """Checks if matrix multiplication is possible."""
    return len(A[0]) == len(B)

def multiply_matrices(A, B):
    """Multiplies two matrices."""
    result = [[sum(A[i][k] * B[k][j] for k in range(len(B))) for j in range(len(B[0]))] for i in range(len(A))]
    return result

def print_matrix(matrix):
    """Prints the matrix."""
    for row in matrix:
        print(" ".join(map(str, row)))

def main():
    # Read dimensions of the first matrix
    rows_A = int(input("Enter the number of rows in the first matrix: "))
    cols_A = int(input("Enter the number of columns in the first matrix: "))

    # Read dimensions of the second matrix
    rows_B = int(input("Enter the number of rows in the second matrix: "))
    cols_B = int(input("Enter the number of columns in the second matrix: "))

    # Read the first matrix
    A = read_matrix(rows_A, cols_A)
    if A is None:
        return

    # Read the second matrix
    B = read_matrix(rows_B, cols_B)
    if B is None:
        return

    # Check if matrix multiplication is possible
    if not can_multiply_matrices(A, B):
        print("Matrix multiplication is not possible. The number of columns in the first matrix must equal the number of rows in the second matrix.")
        return

    # Multiply the matrices
    C = multiply_matrices(A, B)

    # Print the result
    print("The resulting matrix after multiplication is:")
    print_matrix(C)

if __name__ == "__main__":
    main()
```
