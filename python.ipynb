```python
def is_subpartly_perfect(number):
    sum_of_digits = sum(int(digit) for digit in str(number))

    if number % sum_of_digits == 0:
        return True
    else:
        return False

def is_armstrong(number):

    num_digits = len(str(number))


    sum_of_powers = sum(int(digit)**num_digits for digit in str(number))


    if number == sum_of_powers:
        return True
    else:
        return False


if __name__ == "__main__":
    num = int(input("Enter a number: "))

    if is_subpartly_perfect(num):
        print(f"{num} is subpartly perfect.")
    else:
        print(f"{num} is not subpartly perfect.")

    if is_armstrong(num):
        print(f"{num} is an Armstrong number.")
    else:
        print(f"{num} is not an Armstrong number.")
```

```python
def fibonacci_series(n):

    fib1, fib2 = 0, 1


    if n <= 0:
        print("Please enter a positive integer.")
    elif n == 1:
        print(f"Fibonacci series up to {n}:")
        print(fib1)
    else:
        print(f"Fibonacci series up to {n}:")

        print(fib1)
        print(fib2)


        while fib1 + fib2 <= n:
            fib_next = fib1 + fib2
            print(fib_next)
            fib1 = fib2
            fib2 = fib_next


if __name__ == "__main__":
    num = int(input("Enter a number (n): "))
    fibonacci_series(num)
```

```python
import math

def is_krishnamurthy_number(n):

    def factorial(digit):
        return math.factorial(digit)
    str_num = str(n)
    sum_of_factorials = sum(factorial(int(digit)) for digit in str_num)
    return sum_of_factorials == n

number = int(input("Enter a number: "))

if is_krishnamurthy_number(number):
    print(f"{number} is a Krishnamurthy number.")
else:
    print(f"{number} is not a Krishnamurthy number.")
```

```python
def print_geometric_sequence(start, ratio, terms):
    current = start
    for _ in range(terms):
        print(current, end=", " if _ < terms - 1 else "\n")
        current *= ratio

start_term = 2
common_ratio = 3
number_of_terms = 6

print(f"The first {number_of_terms} terms of the geometric sequence are:")
print_geometric_sequence(start_term, common_ratio, number_of_terms)
```

```python
import cmath

def find_roots(a, b, c):
    discriminant = b**2 - 4*a*c
    root1 = (-b + cmath.sqrt(discriminant)) / (2 * a)
    root2 = (-b - cmath.sqrt(discriminant)) / (2 * a)

    return root1, root2

a = float(input("Enter the coefficient a: "))
b = float(input("Enter the coefficient b: "))
c = float(input("Enter the coefficient c: "))

root1, root2 = find_roots(a, b, c)

print(f"The roots of the quadratic equation are {root1} and {root2}.")
```

```python
def print_multiples_of_10(start, end):

    if start % 10 != 0:
        start += 10 - (start % 10)

    for num in range(start, end + 1, 10):
        print(num)

start = int(input("Enter the start of the interval: "))
end = int(input("Enter the end of the interval: "))

print(f"Multiples of 10 between {start} and {end} are:")
print_multiples_of_10(start, end)
```

```python
def find_median(numbers):

    sorted_numbers = sorted(numbers)
    n = len(sorted_numbers)
    if n % 2 == 1:
        median = sorted_numbers[n // 2]
    else:
        middle1 = sorted_numbers[n // 2 - 1]
        middle2 = sorted_numbers[n // 2]
        median = (middle1 + middle2) / 2

    return median

numbers = list(map(float, input("Enter a set of numbers separated by spaces: ").split()))
median = find_median(numbers)
print(f"The median of the set of numbers is {median}.")
```
